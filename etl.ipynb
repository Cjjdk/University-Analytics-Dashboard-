```python





# 1. Convert selected numeric columns
numeric_cols = ["cohort_size", "status", "zip"]
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors="coerce")

# 2. Convert selected datetime columns
date_cols = ["apply_ts", "birthdate", "cohort_start_ts", "cohort_end_ts"]
for col in date_cols:
    data[col] = pd.to_datetime(data[col], errors="coerce")

# Find all categorical columns
cat_cols = data.select_dtypes(include=["category"]).columns

# Convert them to pandas string dtype
data[cat_cols] = data[cat_cols].astype("string")



# Select only string columns
string_cols = data.select_dtypes(include=["string", "object"])



# Check updated datatypes
print(data.dtypes)

# (Optional) Show number of unique types per column
print("\nColumn-wise data type counts:")
print(data.dtypes.value_counts())

# Summary for numeric 
print("\nNumeric Summary:")
print(data[numeric_cols].describe())

# Summary for datetime columns (custom summary)
print("\nDatetime Summary:")
for col in date_cols:
    print(f"\nColumn: {col}")
    print("Min:", data[col].min())
    print("Max:", data[col].max())
    print("Non-Null Count:", data[col].notnull().sum())

# Describe string columns
string_summary = string_cols.describe()

print(string_summary)
```

```python

# Suppose your DataFrame is df
cols = [
    'learner_id', 'country', 'degree', 'institution', 'major',
    'opportunity_id', 'opportunity_name', 'category', 'opportunity_code',
    'cohort_code', 'cohort_start_ts', 'cohort_end_ts', 'cohort_size',
    'apply_ts', 'status', 'email', 'gender', 'birthdate', 'city',
    'state', 'zip'
]

# --- 1. Numeric columns (count numeric 0) ---
numeric_cols = data[cols].select_dtypes(include=["number"]).columns
zero_counts_numeric = (data[numeric_cols] == 0).sum()
total_zeros_numeric = zero_counts_numeric.sum()

print("Zeros in numeric columns:\n", zero_counts_numeric)
print("Total numeric zeros:", total_zeros_numeric)


# --- 2. String/object columns (count "0" as string) ---
string_cols = data[cols].select_dtypes(include=["object"]).columns
zero_counts_string = (data[string_cols].astype(str) == "0").sum()
total_zeros_string = zero_counts_string.sum()

print("\nZeros in string columns:\n", zero_counts_string)
print("Total string '0':", total_zeros_string)


# --- 3. Datetime columns (count "zero dates") ---
# Some datasets store "0", "0000-00-00", or NaT for missing dates
datetime_cols = data[cols].select_dtypes(include=["datetime"]).columns
zero_counts_datetime = ((data[datetime_cols].astype(str) == "0") | 
                        (data[datetime_cols].astype(str) == "0000-00-00")).sum()
total_zeros_datetime = zero_counts_datetime.sum()

print("\nZeros in datetime columns:\n", zero_counts_datetime)
print("Total datetime zeros:", total_zeros_datetime)

```

```python
cols = [
    'learner_id', 'country', 'degree', 'institution', 'major',
    'opportunity_id', 'opportunity_name', 'category', 'opportunity_code',
    'cohort_code', 'cohort_start_ts', 'cohort_end_ts', 'cohort_size',
    'apply_ts', 'status', 'email', 'gender', 'birthdate', 'city',
    'state', 'zip'
]

# count zeros in each column
(data[cols] == 0).sum()
```

```python
data[cols] = data[cols].replace(0, np.nan)  # Replace zeros with NaN for better handling 
```

```python
data.isnull().sum()  # Check for missing values in each column
```

```python
data.dropna(inplace=True)  # Drop rows with any NaN values
```

```python
data.drop_duplicates(subset=['learner_id'], inplace=True)
 # Drop duplicate rows based on all columns
```

```python
# Find rows to drop
bad_rows = data[
    (data['zip'] < 0) | 
    (data['zip'].astype(str).str.len() > 10)
].index

# Drop them inplace
data.drop(index=bad_rows, inplace=True)
```
